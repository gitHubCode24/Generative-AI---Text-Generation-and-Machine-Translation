##Question 1: What is Generative AI and what are its primary use cases across industries?

**Generative AI** refers to a class of artificial intelligence models capable of producing new, original content that resembles existing data but is not a direct copy. Unlike traditional AI that might classify, predict, or analyze existing data, Generative AI creates novel outputs such as text, images, audio, video, and even code.

At its core, Generative AI learns patterns, structures, and styles from vast amounts of training data. Once it has internalized these characteristics, it can then generate new instances that share similar properties. This capability stems from advanced neural network architectures, particularly transformer models and generative adversarial networks (GANs), which allow these models to understand complex relationships within data and synthesize coherent, high-quality content.

**Primary Use Cases Across Industries:**

1.  **Content Creation & Marketing:**
    *   **Text Generation:** Creating articles, blog posts, marketing copy, social media updates, product descriptions, and even entire books. This accelerates content production and helps overcome writer's block.
    *   **Image & Video Generation:** Designing unique images for advertising, creating synthetic data for training other AI models, generating realistic virtual environments, and producing short video clips or animations.
    *   **Music Composition:** Generating background music for videos, creating unique jingles, or assisting artists in composing new pieces.

    *   **Rapid Prototyping:** Generating multiple design variations for products, architectures, or fashion items, allowing designers to iterate quickly.

3.  **Software Development:**
    *   **Code Generation:** Writing code snippets, completing functions, or even generating entire programs based on natural language descriptions, significantly speeding up development cycles.
    *   **Code Refactoring & Optimization:** Suggesting improvements to existing code.

4.  **Customer Service & Experience:**
    *   **Personalized Interactions:** Generating highly personalized responses for chatbots or virtual assistants, improving customer engagement and satisfaction.
    *   **Customized Recommendations:** Creating unique product recommendations or content suggestions tailored to individual user preferences.

5.  **Healthcare:**
    *   **Medical Imaging:** Generating synthetic medical images for training diagnostic AI models, particularly useful where real data is scarce or sensitive.

##Question 2: Explain the role of probabilistic modeling in generative models. How do these models differ from discriminative models?

**The Role of Probabilistic Modeling in Generative Models:**

Probabilistic modeling is fundamental to generative AI. At its core, a generative model learns the underlying probability distribution of the training data. This means it tries to understand how the data was generated. If we denote the input data as \(x\) and any associated labels or attributes as \(y\), a generative model aims to learn the joint probability distribution \(P(x, y)\) or, more generally, \(P(x)\) for unsupervised generation. By learning this joint distribution, the model can:

1.  **Generate New Samples:** Once the model has learned \(P(x)\) or \(P(x, y)\), it can then sample from this distribution to create new data points that resemble the original training data. For example, a generative model trained on images of cats can generate new, unseen images of cats.
2.  **Understand Data Structure:** The probabilistic framework allows the model to capture the complex dependencies and relationships within the data, leading to a deeper understanding of the data's inherent structure.
3.  **Handle Missing Data:** By modeling the entire data distribution, generative models can naturally handle missing data or impute values.
4.  **Anomaly Detection:** Data points that have very low probability under the learned distribution \(P(x)\) can be identified as anomalies or outliers.

Techniques like Variational Autoencoders (VAEs) and Generative Adversarial Networks (GANs) both employ probabilistic concepts, though in different ways. VAEs explicitly learn a latent space and a decoder that models \(P(x|z)\) (the probability of data given a latent variable) and an encoder that models \(P(z|x)\). GANs, on the other hand, implicitly learn the data distribution through a adversarial process between a generator and a discriminator.

**Generative Models vs. Discriminative Models:**

The fundamental difference between generative and discriminative models lies in what probability distribution they model and, consequently, what task they are best suited for:

| Feature             | Generative Models                                     | Discriminative Models                                 |
| :------------------ | :---------------------------------------------------- | :---------------------------------------------------- |
| **Goal**            | Learn the distribution of data, generate new samples. | Learn to distinguish between different classes.       |
| **Modeled Prob.**   | \(P(x, y)\) (joint probability) or \(P(x)\)         | \(P(y|x)\) (conditional probability)                   |
| **Question Answered** | "How was the data generated?" or "What does the data look like?" | "What is the probability of \(y\) given \(x\)?" or "What class does \(x\) belong to?" |
| **Tasks**           | Data generation, density estimation, anomaly detection, imputing missing data. | Classification, regression.                           |
| **Examples**        | Naive Bayes, Latent Dirichlet Allocation (LDA), Hidden Markov Models (HMMs), GANs, VAEs. | Logistic Regression, Support Vector Machines (SVMs), Perceptrons, Traditional Neural Networks (for classification). |
| **Complexity**      | Typically more complex as they model the full data distribution. | Generally simpler as they only model the decision boundary. |

In summary, generative models build a complete statistical model of the data, allowing them to create new, synthetic data points, while discriminative models focus solely on learning the boundary or function that separates different classes for tasks like classification. Both are powerful tools in machine learning, but they serve different purposes based on their underlying probabilistic approaches.

##Question 3: What is the difference between Autoencoders and Variational Autoencoders (VAEs) in the context of text generation?

Both Autoencoders (AEs) and Variational Autoencoders (VAEs) are neural network architectures used for learning efficient data encodings (latent representations) in an unsupervised manner. While they share a common goal of dimensionality reduction and data reconstruction, their mechanisms and suitability for text generation differ significantly.

**Autoencoders (AEs):**

An Autoencoder consists of two main parts:

1.  **Encoder:** Maps an input \(x\) to a lower-dimensional latent representation (code) \(z\). Mathematically, \(z = f(x)\).
2.  **Decoder:** Maps the latent representation \(z\) back to a reconstruction of the original input \(x'\). Mathematically, \(x' = g(z)\).

The network is trained to minimize the reconstruction error between the original input \(x\) and its reconstruction \(x'\). The goal is to learn a compact and meaningful representation of the input data.

**In the context of text generation:**

*   **How they work:** For text, the input \(x\) would be a text sequence (e.g., represented as word embeddings or one-hot vectors). The encoder compresses this sequence into a fixed-size vector \(z\), and the decoder attempts to reconstruct the original text from \(z\).
*   **Limitation for generation:** The main drawback of standard AEs for *generation* is that their latent space is not typically continuous or well-structured. If you sample a random \(z\) from a region of the latent space where no training data was encoded, the decoder might produce meaningless or corrupted text. AEs are good at reconstructing data seen during training but struggle to *generate* novel, coherent samples by sampling from their latent space.
*   **Use cases:** Primarily used for dimensionality reduction, feature learning, and denoising text data (e.g., correcting typos).

**Variational Autoencoders (VAEs):**

VAEs are a generative model built upon the Autoencoder framework, but with a crucial probabilistic twist. Instead of the encoder producing a single fixed latent vector \(z\), it produces parameters for a probability distribution (typically a Gaussian distribution) in the latent space.

1.  **Encoder:** Maps an input \(x\) to a mean vector \(\mu\) and a standard deviation (or log-variance) vector \(\sigma\) for a multivariate Gaussian distribution. So, \(z \sim N(\mu(x), \sigma(x))\).
2.  **Reparameterization Trick:** To allow backpropagation through the sampling process, a sample \(z\) is drawn from this distribution using \(z = \mu + \sigma \cdot \epsilon\), where \(\epsilon \sim N(0, 1)\).
3.  **Decoder:** Maps the sampled latent vector \(z\) to a reconstruction of the original input \(x'\), similar to a standard AE.

VAEs are trained with a composite loss function:

*   **Reconstruction Loss:** Ensures \(x'\) is similar to \(x\) (like AEs).
*   **KL Divergence Loss:** Forces the learned latent distribution \(N(\mu(x), \sigma(x))\) to be close to a prior distribution (usually a standard normal distribution \(N(0, 1)\)). This regularizes the latent space, making it smooth and continuous.

**In the context of text generation:**

*   **How they work:** For text, the encoder learns to map input text to a distribution over latent space. The decoder then samples from this latent distribution and generates text. Because the latent space is forced to be continuous and well-structured by the KL divergence term, sampling a random \(z\) from the prior (e.g., \(N(0, 1)\)) and passing it through the decoder is more likely to yield coherent and meaningful text.
*   **Advantages for generation:** The regularized latent space allows VAEs to act as true generative models. By sampling from any point in the latent space (which is continuous), the decoder can generate new, diverse, and plausible text sequences. It enables interpolation between different text samples by navigating the latent space.
*   **Use cases:** Text generation (e.g., generating sentences with specific styles or topics), text style transfer, controllable text generation, and learning disentangled representations of text attributes.

**Key Differences Summarized for Text Generation:**

| Feature           | Autoencoder (AE)                                    | Variational Autoencoder (VAE)                                    |
| :---------------- | :-------------------------------------------------- | :--------------------------------------------------------------- |
| **Latent Space**  | Arbitrary, not necessarily continuous or structured. | Continuous, structured, and regularized by a prior distribution. |
| **Encoder Output**| A single latent vector \(z\).                     | Parameters (mean \(\mu\) and variance \(\sigma\)) of a distribution for \(z\). |
| **Generation Cap.** | Poor for generating *novel* samples; good for reconstruction. | Good for generating *novel* and diverse samples by sampling from latent space. |
| **Loss Function** | Reconstruction error only.                          | Reconstruction error + KL divergence (regularization).         |
| **Probabilistic** | No inherent probabilistic model of data.            | Explicitly models the data as being generated from a latent probability distribution. |

In essence, while an Autoencoder learns to compress and reconstruct text efficiently, a Variational Autoencoder learns a probabilistic latent space that enables it to *generate* new, diverse, and coherent text by sampling from that learned distribution.

##Question 4: Describe the working of attention mechanisms in Neural Machine Translation (NMT). Why are they critical?

**Working of Attention Mechanisms in Neural Machine Translation (NMT):**

Before attention, traditional Sequence-to-Sequence (Seq2Seq) models for NMT (using an Encoder-Decoder architecture) struggled with long sentences. The encoder would compress the entire source sentence into a single fixed-size context vector, and the decoder would generate the target sentence solely based on this vector. This often led to information loss for longer sequences, making it difficult for the decoder to accurately translate the entire input.

Attention mechanisms were introduced to overcome this limitation. Instead of compressing the entire source sentence into a single vector, attention allows the decoder to "look back" at different parts of the source sentence (encoder hidden states) at each step of generating a target word. This process can be broken down as follows:

1.  **Encoder:** The encoder (typically a Recurrent Neural Network like an LSTM or GRU) processes the source sentence word by word, generating a sequence of hidden states. Each hidden state \(h_i\) for source word \(i\) contains information about that word and its context in the source sentence.

2.  **Decoder (with Attention):** At each step when the decoder is about to generate a target word:
    *   **Context Query:** The decoder's current hidden state (or a previous hidden state) acts as a query to determine which parts of the source sentence are most relevant for generating the next target word.
    *   **Alignment Scores:** An alignment model (often a feed-forward neural network) computes a score for how well each encoder hidden state \(h_j\) aligns with the current decoder state. These scores indicate the importance or relevance of each source word to the current target word being generated.
    *   **Softmax Normalization:** These alignment scores are then passed through a softmax function to produce a set of weights (attention weights). These weights sum up to 1 and represent the probability distribution over the source words, indicating how much "attention" should be given to each source word.
    *   **Context Vector Creation:** A weighted sum of the encoder's hidden states is computed, using the attention weights. This weighted sum forms a new context vector for the current decoding step. This context vector is dynamic, changing at each step to reflect the most relevant parts of the source sentence.
    *   **Output Generation:** The decoder then combines this context vector with its current hidden state to predict the next target word. This allows the decoder to focus on specific, relevant parts of the source sentence as it generates each word of the translation.

**Why Attention Mechanisms are Critical:**

Attention mechanisms are critical in NMT for several reasons:

1.  **Overcoming Fixed-Size Context Vector Bottleneck:** This was the primary motivation. Attention alleviates the problem of compressing all information into a single vector, which often led to information loss, especially for long sentences. It allows relevant information to flow directly from the encoder to the decoder at each step.

2.  **Handling Long-Range Dependencies:** By directly attending to all parts of the input sequence, attention mechanisms are much better at capturing long-range dependencies and translating sentences accurately, regardless of their length.

3.  **Improved Performance and Accuracy:** Attention models significantly boosted the performance of NMT systems, leading to more fluid, grammatically correct, and semantically accurate translations compared to non-attention-based models.

4.  **Interpretability:** Attention weights provide a degree of interpretability. By visualizing these weights, one can see which source words the model is focusing on when generating a particular target word. This offers insights into the translation process and helps in debugging or understanding model behavior.

5.  **Handling Word Order Differences:** Languages often have different word orders. Attention allows the model to flexibly align source and target words, even if their positions are not directly corresponding, making it robust to varying grammatical structures.

6.  **Foundation for Transformers:** The success of attention mechanisms directly led to the development of the Transformer architecture, which completely relies on self-attention and cross-attention mechanisms. Transformers have become the dominant architecture in NMT and many other NLP tasks, proving the fundamental importance of attention.

In summary, attention mechanisms transformed NMT by providing a dynamic way for the decoder to access relevant source information, leading to substantial improvements in translation quality, handling of long sentences, and providing valuable interpretability.

##Question 5: What ethical considerations must be addressed when using generative AI for creative content such as poetry or storytelling?

### Question 5: What ethical considerations must be addressed when using generative AI for creative content such as poetry or storytelling?

When using generative AI for creative content like poetry or storytelling, several significant ethical considerations must be addressed to ensure responsible and fair development and deployment:

1.  **Authorship and Attribution:**
    *   **Who is the author?** When an AI generates content, the question of true authorship becomes blurred. Is it the AI, the programmer, the data scientists who curated the training data, or the user who prompted it? Clear guidelines are needed.
    *   **Attribution:** Should AI-generated content always be disclosed? Hiding AI involvement can deceive audiences and undermine trust. Proper attribution ensures transparency and respect for human creativity.
    *   **Intellectual Property Rights:** Who owns the copyright for AI-generated poems, stories, or other creative works? Current IP laws are often based on human authorship, creating legal ambiguities.

2.  **Originality and Authenticity:**
    *   **Plagiarism/Derived Work:** Generative AI learns from vast datasets of existing human-created content. There's a risk that its output might inadvertently replicate or closely mimic existing works, leading to accusations of plagiarism or lack of originality.
    *   **Authenticity of Human Experience:** Creative content often reflects unique human experiences, emotions, and perspectives. When AI generates such content, it can lack the genuine emotional depth or lived experience that resonates with human readers, potentially devaluing human artistry.

3.  **Bias and Stereotypes:**
    *   **Reinforcement of Bias:** AI models are trained on existing text data, which often contains societal biases, stereotypes, and harmful narratives. Generative AI can perpetuate and amplify these biases, leading to the creation of content that is discriminatory, offensive, or reinforces harmful stereotypes against certain groups.
    *   **Lack of Diversity:** If training data lacks diverse voices and perspectives, the AI's creative output might be limited in scope, reflecting only dominant cultural narratives and excluding marginalized ones.

4.  **Misinformation and Manipulation (Deepfakes of Text):**
    *   **Fabrication of Narratives:** Generative AI can produce highly convincing fictional stories, speeches, or articles that could be used to spread misinformation, create propaganda, or manipulate public opinion if presented as genuine human-authored content.
    *   **Reputation Damage:** AI can be used to generate malicious or defamatory content about individuals or organizations, harming reputations and leading to real-world consequences.

5.  **Economic Impact on Human Creatives:**
    *   **Job Displacement:** The ability of AI to rapidly produce creative content raises concerns about the economic viability of human writers, poets, and storytellers. If AI can generate content faster and cheaper, it might reduce demand for human-created works.
    *   **Fair Compensation:** If AI tools are used to assist human creatives, how should compensation be structured to fairly acknowledge the AI's contribution without devaluing the human's role?

6.  **Erosion of Critical Thinking and Creativity:**
    *   **Over-reliance:** An over-reliance on AI for creative tasks might stifle human creativity, critical thinking, and the development of unique artistic skills.
    *   **Passive Consumption:** If audiences are constantly exposed to AI-generated content without disclosure, it could erode their ability to distinguish between human and machine creativity, potentially leading to a more passive consumption of art.

7.  **Data Privacy and Consent:**
    *   **Training Data Ethics:** If proprietary or copyrighted works are used in training data, questions arise about proper consent from the original creators and fair compensation for their intellectual property.

Addressing these ethical considerations requires a multi-faceted approach involving responsible AI development, transparent disclosure practices, legal frameworks, public education, and ongoing societal dialogue to ensure that generative AI enhances human creativity rather than undermining it.

##Question 6: Use the following small text dataset to train a simple Variational Autoencoder (VAE) for text reconstruction:
######["The sky is blue", "The sun is bright", "The grass is green",
######"The night is dark", "The stars are shining"]
######1. Preprocess the data (tokenize and pad the sequences).
######2. Build a basic VAE model for text reconstruction.
######3. Train the model and show how it reconstructs or generates similar sentences.
######Include your code, explanation, and sample outputs.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np

# Data
texts = [
    "The sky is blue", "The sun is bright", "The grass is green",
    "The night is dark", "The stars are shining"
]

# Tokenization
tokenizer = keras.preprocessing.text.Tokenizer(num_words=None, oov_token="<unk>")
tokenizer.fit_on_texts(texts)
word_index = tokenizer.word_index
vocabulary_size = len(word_index) + 1
sequences = tokenizer.texts_to_sequences(texts)
max_sequence_length = max(len(seq) for seq in sequences)
padded_sequences = keras.preprocessing.sequence.pad_sequences(
    sequences, maxlen=max_sequence_length, padding='post'
)
X_train = np.array(padded_sequences)

# VAE Components
class Sampling(layers.Layer):
    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.keras.backend.random_normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon

embedding_dim = 16
latent_dim = 8
intermediate_dim = 32

encoder_inputs = keras.Input(shape=(max_sequence_length,))
x = layers.Embedding(vocabulary_size, embedding_dim)(encoder_inputs)
x = layers.LSTM(intermediate_dim, return_sequences=False)(x)
z_mean = layers.Dense(latent_dim, name="z_mean")(x)
z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)
z = Sampling()([z_mean, z_log_var])
encoder = keras.Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")

latent_inputs = keras.Input(shape=(latent_dim,))
x = layers.RepeatVector(max_sequence_length)(latent_inputs)
x = layers.LSTM(intermediate_dim, return_sequences=True)(x)
decoder_outputs = layers.TimeDistributed(layers.Dense(vocabulary_size, activation="softmax"))(x)
decoder = keras.Model(latent_inputs, decoder_outputs, name="decoder")

class VAE(keras.Model):
    def __init__(self, encoder, decoder, **kwargs):
        super().__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        self.total_loss_tracker = keras.metrics.Mean(name="total_loss")
        self.reconstruction_loss_tracker = keras.metrics.Mean(
            name="reconstruction_loss"
        )
        self.kl_loss_tracker = keras.metrics.Mean(name="kl_loss")

    @property
    def metrics(self):
        return [
            self.total_loss_tracker,
            self.reconstruction_loss_tracker,
            self.kl_loss_tracker,
        ]

    def train_step(self, data):
        x = data
        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(x)
            reconstruction = self.decoder(z)
            reconstruction_loss = keras.losses.SparseCategoricalCrossentropy(
                from_logits=False, reduction=tf.keras.losses.Reduction.NONE
            )(tf.cast(x, tf.int32), reconstruction)
            reconstruction_loss = tf.reduce_mean(reconstruction_loss)
            kl_loss = -0.5 * (1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var))
            kl_loss = tf.reduce_mean(tf.reduce_sum(kl_loss, axis=1))
            total_loss = reconstruction_loss + kl_loss

        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))
        self.total_loss_tracker.update_state(total_loss)
        self.reconstruction_loss_tracker.update_state(reconstruction_loss)
        self.kl_loss_tracker.update_state(kl_loss)
        return {
            "loss": self.total_loss_tracker.result(),
            "reconstruction_loss": self.reconstruction_loss_tracker.result(),
            "kl_loss": self.kl_loss_tracker.result(),
        }

    def call(self, inputs):
        z_mean, z_log_var, z = self.encoder(inputs)
        reconstruction = self.decoder(z)
        return reconstruction

# Instantiate and Compile VAE
vae = VAE(encoder, decoder)
vae.compile(optimizer=keras.optimizers.Adam())

# Train Model
print("Training VAE...")
vae.fit(X_train, epochs=200, batch_size=len(texts))
print("Training complete.")

# Function to convert sequence back to text
def sequence_to_text(sequence, tokenizer):
    reverse_word_index = dict([(value, key) for (key, value) in tokenizer.word_index.items()])
    text_tokens = [reverse_word_index.get(token, '') for token in sequence if token != 0]
    return ' '.join(text_tokens)

# Reconstruction Examples
print("\n--- Reconstruction Examples ---")
reconstructed_outputs = vae.predict(X_train)
for i, original_text in enumerate(texts):
    predicted_sequence = np.argmax(reconstructed_outputs[i], axis=-1)
    reconstructed_text = sequence_to_text(predicted_sequence, tokenizer)
    print(f"Original: '{original_text}'")
    print(f"Reconstructed: '{reconstructed_text}'\n")

# Generation Examples
print("\n--- Generation Examples ---")
num_generations = 5
for _ in range(num_generations):
    random_latent_vector = tf.random.normal(shape=(1, latent_dim))
    generated_sequence_probs = decoder.predict(random_latent_vector)
    generated_sequence = np.argmax(generated_sequence_probs[0], axis=-1)
    generated_text = sequence_to_text(generated_sequence, tokenizer)
    print(f"Generated: '{generated_text}'")
```

##Question 8: Implement a simple attention-based encoder-decoder model for English-to-Spanish translation using Tensorflow or PyTorch.

```python
import torch
import torch.nn as nn
import torch.optim as optim

device = "cuda" if torch.cuda.is_available() else "cpu"

# ----- Encoder -----
class Encoder(nn.Module):
    def __init__(self, vocab, emb, hid):
        super().__init__()
        self.emb = nn.Embedding(vocab, emb)
        self.rnn = nn.LSTM(emb, hid, batch_first=True)

    def forward(self, x):
        x = self.emb(x)
        out, (h, c) = self.rnn(x)
        return out, h, c

# ----- Attention -----
class Attention(nn.Module):
    def __init__(self, hid):
        super().__init__()
        self.fc = nn.Linear(hid*2, hid)
        self.v = nn.Linear(hid, 1, bias=False)

    def forward(self, h, enc_out):
        h = h[-1].unsqueeze(1).repeat(1, enc_out.size(1), 1)
        energy = torch.tanh(self.fc(torch.cat((h, enc_out), 2)))
        return torch.softmax(self.v(energy).squeeze(2), dim=1)

# ----- Decoder -----
class Decoder(nn.Module):
    def __init__(self, vocab, emb, hid, attn):
        super().__init__()
        self.emb = nn.Embedding(vocab, emb)
        self.rnn = nn.LSTM(emb+hid, hid, batch_first=True)
        self.fc = nn.Linear(hid*2, vocab)
        self.attn = attn

    def forward(self, x, h, c, enc_out):
        x = self.emb(x.unsqueeze(1))
        w = self.attn(h, enc_out).unsqueeze(1)
        ctx = torch.bmm(w, enc_out)
        out, (h, c) = self.rnn(torch.cat((x, ctx), 2), (h, c))
        pred = self.fc(torch.cat((out.squeeze(1), ctx.squeeze(1)), 1))
        return pred, h, c

# ----- Seq2Seq -----
class Seq2Seq(nn.Module):
    def __init__(self, enc, dec):
        super().__init__()
        self.enc, self.dec = enc, dec

    def forward(self, src, trg):
        enc_out, h, c = self.enc(src)
        out = torch.zeros(trg.size(0), trg.size(1), dec.fc.out_features).to(device)
        x = trg[:, 0]

        for t in range(1, trg.size(1)):
            pred, h, c = self.dec(x, h, c, enc_out)
            out[:, t] = pred
            x = pred.argmax(1)

        return out

# ----- Init -----
VOCAB = 5000
EMB, HID = 128, 256

attn = Attention(HID)
enc = Encoder(VOCAB, EMB, HID)
dec = Decoder(VOCAB, EMB, HID, attn)
model = Seq2Seq(enc, dec).to(device)

# ----- Dummy Data (replace with real data) -----
src = torch.randint(0, VOCAB, (32, 10)).to(device)
trg = torch.randint(0, VOCAB, (32, 10)).to(device)

# ----- Train Step -----
opt = optim.Adam(model.parameters())
loss_fn = nn.CrossEntropyLoss()

out = model(src, trg)

loss = loss_fn(out[:,1:].reshape(-1, VOCAB), trg[:,1:].reshape(-1))
loss.backward()
opt.step()

print("Loss:", loss.item())
```

```python
from transformers import pipeline

# Load GPT-2
generator = pipeline("text-generation", model="gpt2")

# Given dataset (used as style reference)
dataset = [
    "Roses are red, violets are blue,",
    "Sugar is sweet, and so are you.",
    "The moon glows bright in silent skies,",
    "A bird sings where the soft wind sighs."
]

# Prompt (simulating fine-tuning via pattern imitation)
prompt = (
    "Write a short 2-4 line poem similar in style:\n"
    "Roses are red, violets are blue,\n"
    "Sugar is sweet, and so are you.\n"
    "The moon glows bright in silent skies,\n"
    "A bird sings where the soft wind sighs.\n\n"
    "New poem:\n"
)

# Generate poem
output = generator(prompt, max_length=80, num_return_sequences=1)

# Extract generated poem
generated_text = output[0]['generated_text'].split("New poem:")[-1].strip()

print("=== PROMPT USED ===")
print(prompt)

print("\n=== GENERATED POEM ===")
print(generated_text)
```

##Question 10: Imagine you are building a creative writing assistant for a publishing company. The assistant should generate story plots and character descriptions using Generative AI. Describe how you would design the system, including model selection, training data, bias mitigation, and evaluation methods. Explain the real-world challenges you might face.


**1. Model Selection:**
*   **Story Plots:** Large Language Models (LLMs) like GPT-3/4 or fine-tuned variants of models like T5 or LLaMA, capable of long-form text generation and understanding complex narrative structures.
*   **Character Descriptions:** Similar LLMs, potentially fine-tuned on datasets rich in character profiles and psychological traits.

**2. Training Data:**
*   **Diverse Literary Corpus:** A vast collection of novels, short stories, screenplays, and character encyclopedias, covering multiple genres, cultures, and writing styles. Ensure variety in plot structures, character archetypes, and narrative voices.
*   **Structured Data:** Potentially use datasets of plot summaries, character sheets, and story outlines with associated tags to enable more controllable generation.

**3. Bias Mitigation:**
*   **Data Curation:** Carefully filter training data to reduce overrepresentation of stereotypes and harmful biases. Actively seek out and include diverse literary works.
*   **Bias Detection Tools:** Employ tools to identify and quantify biases in generated output (e.g., gender, racial, cultural stereotypes).
*   **Controlled Generation:** Implement mechanisms (e.g., specific prompts, ethical filters, reinforcement learning with human feedback (RLHF)) to guide the model towards generating inclusive and unbiased content.

**4. Evaluation Methods:**
*   **Human-in-the-Loop:** Publishing professionals (editors, writers) evaluate generated plots and characters for creativity, coherence, originality, adherence to prompts, and absence of bias.
*   **Automated Metrics:** Use metrics for text quality (e.g., perplexity, BLEU for fluency/coherence if comparing to reference texts), but these are less effective for open-ended creativity.
*   **Diversity Metrics:** Quantify the range and uniqueness of generated outputs.
*   **User Feedback:** Collect structured feedback from writers using the assistant.

**5. Real-World Challenges:**
*   **Maintaining Originality:** Avoiding plagiarism or repetitive outputs from the training data.
*   **Lack of Genuine Creativity:** AI's outputs may lack true innovation, emotional depth, or a unique "voice" that distinguishes human writers.
*   **Bias Persistence:** Despite mitigation efforts, subtle biases from vast training data can still manifest.
*   **Scalability for Human Review:** Evaluating a large volume of generated content by human experts is time-consuming and expensive.
*   **Ethical & IP Concerns:** Questions of authorship, copyright, and the displacement of human jobs.
*   **Controllability:** Directing the AI to generate specific plot twists or character arcs without losing coherence or creativity.
